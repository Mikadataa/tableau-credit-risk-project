```python
import pandas as pd
import numpy as np

#Load dataset
df = pd.read_csv("../data/UCI_Credit_Card.csv")

#Preview
df.head()

```

```python
#Checking shape and column names
df.shape
df.columns

# Checking shape and column names
# - df.shape: tells us the number of rows (clients) and columns (variables).
# - df.columns: lists all column names so we understand what data is available.
# This helps verify the dataset was loaded correctly and gives us a sense of size.
```

```python
# Checking for missing values
# - Important to identify any NaN/null values before analysis.
# - If found, we would decide whether to impute or drop them.
df.isnull().sum()

# Checking data types
# - Ensures numeric columns are numeric, categorical are categorical.
# - If not, we will convert them (e.g., Gender, Education).
df.dtypes

# Summary statistics
# - Gives min, max, mean, std, quartiles.
# - Useful to detect anomalies (e.g., negative ages, outlier credit limits).
df.describe()
```

```python
#Renaming Columns for Readability

df.rename(columns={
    'LIMIT_BAL': 'Credit_Limit',
    'SEX': 'Gender',
    'EDUCATION': 'Education',
    'MARRIAGE': 'Marital_Status',
    'AGE': 'Age',
    'PAY_0': 'Pay_Sep',
    'PAY_2': 'Pay_Aug',
    'PAY_3': 'Pay_Jul',
    'PAY_4': 'Pay_Jun',
    'PAY_5': 'Pay_May',
    'PAY_6': 'Pay_Apr',
    'default.payment.next.month': 'Default_Next_Month'
}, inplace=True)

```

```python
#Map categories
# Gender
df['Gender'] = df['Gender'].map({1: "Male", 2: "Female"})

# Education
edu_map = {1: "Graduate", 2: "University", 3: "High School", 4: "Others", 5: "Unknown", 6: "Unknown"}
df['Education'] = df['Education'].replace(edu_map)

# Marital Status
mar_map = {1: "Married", 2: "Single", 3: "Others"}
df['Marital_Status'] = df['Marital_Status'].replace(mar_map)
```

```python
#Creating new features


# 1) Make sure the "latest status" column exists under a known name
if 'Pay_Sep' not in df.columns and 'PAY_0' in df.columns:
    df = df.rename(columns={'PAY_0': 'Pay_Sep'})

# 2) Create Delinquency_Stage safely
def map_stage(x):
    if pd.isna(x) or x <= 0:   # -1, 0, or NaN -> Current
        return "Current"
    elif x <= 2:
        return "Early"
    elif x <= 4:
        return "Mid"
    else:
        return "Late"

df['Delinquency_Stage'] = df['Pay_Sep'].apply(map_stage)

# 3) Repayment_Ratio without divide-by-zero
bill_cols = [f'BILL_AMT{i}' for i in range(1,7)]
pay_cols  = [f'PAY_AMT{i}'  for i in range(1,7)]

bills_sum = df[bill_cols].sum(axis=1, min_count=1)
pays_sum  = df[pay_cols].sum(axis=1, min_count=1)
bills_safe = bills_sum.mask(bills_sum <= 0, np.nan)

df['Repayment_Ratio'] = pays_sum / bills_safe

# (Optional) band the ratio for easier Tableau visuals
bins   = [-np.inf, 0, 0.25, 0.75, 1.0, np.inf]
labels = ['0%', '0–25%', '25–75%', '75–100%', '>100%']
df['Repayment_Band'] = pd.cut(df['Repayment_Ratio'], bins=bins, labels=labels)

# sanity checks
print(df[['Pay_Sep','Delinquency_Stage','Repayment_Ratio','Repayment_Band']].head())
```

Delinquency -> a borrower is late on payments.

#Stages show how overdue the account is:
- Current → Account is up to date (paid on time).
- Early → Recently overdue (e.g., 1–2 months late).
- Mid → Medium risk, more overdue (e.g., 3–4 months late).
- Late → Seriously overdue (5+ months late). Often considered “bad debt” if not collected.

In Our Dataset:
The PAY_0 column (and PAY_2–PAY_6) show repayment status:
-1 = Paid duly (no delay)
0 = Paid on time
1 = Delay 1 month
2 = Delay 2 months
… up to 9 = Delay 9+ months

👉 We use PAY_0 (September, latest month) to define the current stage
